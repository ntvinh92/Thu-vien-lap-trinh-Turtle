```python
import turtle
import math
t = turtle.Turtle()
t.speed(10)
turtle.bgcolor('SkyBlue')

nha_x = 300
nha_y = 300
nha_tuong = 150
nha_san_mai = 120

cay_x = 200
cay_y = 350
cay_than = 60
cay_goc = 40
cay_canh = 80

nang_x = 150
nang_y = 200
tia_nang1 = 100
tia_nang2 = 80
```

```python
### Vẽ ngôi nhà
t.penup()
t.goto(-nha_x, -nha_y)
t.pendown()
t.fillcolor('blue')
t.begin_fill()
for i in range (2):
    t.fd(nha_san_mai)
    t.lt(90)
    t.fd(nha_tuong)
    t.lt(90)
t.end_fill()

t.penup()
t.goto(-nha_x, -(nha_y - nha_tuong))
t.pendown()
t.lt(60)
t.fillcolor('pink')
t.begin_fill()
t.fd(nha_san_mai)
t.rt(120)
t.fd(nha_san_mai)
t.end_fill()

t.fillcolor('orange')
t.begin_fill()
for i in range (4):
    t.lt(90)
    t.fd(nha_san_mai)
t.end_fill()

t.fillcolor('yellow')
t.begin_fill()
t.rt(30)
t.fd(nha_tuong)
t.lt(120)
t.fd(nha_san_mai)
t.lt(60)
t.fd(nha_tuong)
t.end_fill()

t.penup()
t.goto(-(nha_x - nha_san_mai/4), -nha_y) 
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.fd(nha_tuong/2)
t.rt(90)
t.fd(nha_san_mai/2)
t.rt(90)
t.fd(nha_tuong/2)
t.end_fill()

t.penup()
t.goto(-(nha_x - nha_san_mai - (nha_san_mai/3)/math.sqrt(2)), -(nha_y - nha_tuong)) 
t.pendown()
t.fillcolor('brown')
t.begin_fill()
t.fd(nha_tuong/2)
t.lt(120)
t.fd(nha_san_mai/2)
t.lt(60)
t.fd(nha_tuong/2)
t.lt(120)
t.fd(nha_san_mai/2)
t.end_fill()
```

```python
### Vẽ cây
# Gốc cây
t.penup()
t.goto(cay_x, -cay_y)
t.pendown()
t.fillcolor('brown')
t.begin_fill()
t.lt(150)
for i in range (2):
    t.fd(cay_goc)
    t.lt(90)
    t.fd(cay_than)
    t.lt(90)
t.end_fill()

# Cành cây 1
t.penup()
t.goto(cay_x + cay_goc/2, -(cay_y - cay_than))
t.pendown()
t.fillcolor('lightgreen')
t.begin_fill()
t.fd(cay_canh)
t.lt(135)
t.fd(cay_canh*math.sqrt(2))
t.lt(90)
t.fd(cay_canh*math.sqrt(2))
t.lt(135)
t.fd(cay_canh)

# Cành cây 2
t.penup()
t.goto(cay_x + cay_goc/2, -(cay_y - cay_than - cay_canh))
t.pendown()
t.fd(cay_canh)
t.lt(135)
t.fd(cay_canh*math.sqrt(2))
t.lt(90)
t.fd(cay_canh*math.sqrt(2))
t.lt(135)
t.fd(cay_canh)

# Cành cây 3
t.penup()
t.goto(cay_x + cay_goc/2, -(cay_y - cay_than - 2*cay_canh))
t.pendown()
t.fd(cay_canh)
t.lt(135)
t.fd(cay_canh*math.sqrt(2))
t.lt(90)
t.fd(cay_canh*math.sqrt(2))
t.lt(135)
t.fd(cay_canh)
t.end_fill()
```

```python
### Vẽ mặt trời
# Tia nắng
t.penup()
t.goto(nang_x, nang_y)
t.pendown()
for i in range (2):
    t.fd(tia_nang1)
    t.bk(tia_nang1)
    t.lt(180)
    t.fd(tia_nang1)
    t.bk(tia_nang1)
    t.rt(45)
    t.fd(tia_nang2)
    t.bk(tia_nang2)
    t.lt(180)
    t.fd(tia_nang2)
    t.bk(tia_nang2)
    t.rt(45)

# Mặt trời
t.goto(nang_x, nang_y + tia_nang1/2)
t.fillcolor('yellow')
t.begin_fill()
t.circle(50)
t.end_fill()
```
