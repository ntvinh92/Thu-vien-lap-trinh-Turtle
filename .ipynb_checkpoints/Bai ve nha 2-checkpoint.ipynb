```python
import turtle
t = turtle.Turtle()
t.pensize(2)

# Hinh vuông 1
t.forward(50)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(100)
t.right(90)
t.forward(50)

# Hinh vuông 2
t.right(90)
t.forward(50)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(50)

turtle.done()
```
