```python
import turtle
t = turtle.Turtle()

# Hình vuông 1
t.right(18)
t.forward(50)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(50)

# Hình vuông 2, 3, 4, 5
for i in range(4):
    t.right(72)
    t.forward(50)
    t.left(90)
    t.forward(100)
    t.left(90)
    t.forward(100)
    t.left(90)
    t.forward(100)
    t.left(90)
    t.forward(50)

turtle.done()
```
