```python
### Import thu vien Turtle
import turtle
```

```python
### Khoi tao doi tuong ve
t = turtle.Turtle()
### Ve hinh vuong
t.pensize(10)
t.fillcolor('gray')
t.begin_fill()
t.pencolor('red')
t.forward(100)
t.left(90)
t.pencolor('green')
t.forward(100)
t.left(90)
t.pencolor('blue')
t.forward(100)
t.left(90)
t.pencolor('yellow')
t.forward(100)
t.left(90)
t.end_fill()
### Ket thuc ve hinh vuong
turtle.done()
```
