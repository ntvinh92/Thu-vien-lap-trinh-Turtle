```python
import turtle
import math
t = turtle.Turtle()
canh = 100
canh1 = canh*(math.sqrt(2)-1)/2

# Hình vuông 1
for i in range(4):
    t.forward(canh)
    t.left(90)

# Hình vuông 2
t.penup()
t.goto(canh/2, -canh1)
t.pendown()
t.lt(45)
for i in range(4):
    t.forward(canh)
    t.left(90)
    
turtle.done()
```
